## Assignment Ensemble Learning (Theoretical)

1. Can we use Bagging for regression problems?

- Yes, Bagging can be used for regression problems. This is achieved through a Bagging Regressor.



2. What is the difference between multiple model training and single model training?

- Single Model Training: Involves training one model (e.g., a single Decision Tree or SVM) on the entire dataset. The model learns a single set of parameters or rules.

- Multiple Model Training (Ensemble Learning): Involves training several models (called base estimators) and combining their predictions to make a final prediction. This often leads to improved robustness and accuracy compared to a single model. Examples include Bagging, Boosting, and Stacking.

3. Explain the concept of feature randomness in Random Forest.

- Feature randomness in Random Forest refers to the process where, at each split in a decision tree, only a random subset of features is considered for finding the best split.  This decorrelates the trees in the forest, making them more diverse and reducing the variance of the overall model.

4. What is OOB (Out-of-Bag) Score?

- The Out-of-Bag (OOB) Score is a way to estimate the generalization error of a Bagging-based ensemble model, like Random Forest, without the need for a separate validation set.  During bootstrap sampling, approximately one-third of the data is left out of each bootstrap sample. These "out-of-bag" samples can then be used to evaluate the performance of the trees that were not trained on them.

5. How can you measure the importance of features in a Random Forest model?

- Feature importance in a Random Forest model can be measured based on how much each feature contributes to reducing impurity (e.g., Gini impurity for classification, or MSE for regression) across all the trees in the forest. Features that lead to larger reductions in impurity when used for splitting are considered more important.  These importance scores are typically normalized so that their sum is 1.

6. Explain the working principle of a Bagging Classifier.

-  A Bagging Classifier works by training multiple base classifiers (often Decision Trees) on different bootstrap samples of the original dataset.  Each base classifier is trained independently. For prediction, the Bagging Classifier aggregates the predictions of all base classifiers, typically by taking a majority vote for classification tasks. This process helps reduce variance and overfitting.


7. How do you evaluate a Bagging Classifier's performance?

- The performance of a Bagging Classifier can be evaluated using standard metrics such as:

- Accuracy: For classification problems.
- Precision, Recall, F1-score: Especially important for imbalanced datasets.
- AUC-ROC score: For binary classification, measuring the ability to distinguish between classes.
- OOB (Out-of-Bag) Score: A convenient way to estimate performance internally without a separate validation set.
- Cross-validation: A robust method to estimate generalization performance.

8. How does a Bagging Regressor work?
- A Bagging Regressor works similarly to a Bagging Classifier, but for regression tasks. It trains multiple base regressors (e.g., Decision Trees) on different bootstrap samples of the training data. For new predictions, it averages the predictions from all the individual base regressors.  This averaging helps to reduce the variance of the overall model.

9. What is the main advantage of ensemble techniques?

- The main advantage of ensemble techniques is that they typically improve the overall prediction performance (accuracy for classification, lower error for regression) and robustness of a model.  They achieve this by combining the strengths of multiple individual models, often reducing bias, variance, or both.

10. What is the main challenge of ensemble methods?

- The main challenge of ensemble methods is their increased computational cost and complexity compared to single models.  Training and storing multiple models can be time-consuming and resource-intensive. They can also be harder to interpret than single models.

11. Explain the key idea behind ensemble techniques.

- The key idea behind ensemble techniques is to combine the predictions of multiple individual models (often referred to as "base learners" or "weak learners") to achieve a more robust and accurate overall prediction than any single model could achieve on its own.  The principle is that by aggregating diverse models, the errors of individual models can be compensated for, leading to better generalization.

12. What is a Random Forest Classifier?

- A Random Forest Classifier is an ensemble learning method that builds a "forest" of decision trees. It combines the bagging technique with "feature randomness." Each tree in the forest is trained on a bootstrap sample of the data, and at each split, only a random subset of features is considered. For classification, the final prediction is made by majority voting among the trees.

13. What are the main types of ensemble techniques?

- The main types of ensemble techniques are:
    -Bagging (Bootstrap Aggregating): Trains multiple base models independently on bootstrap samples of the data and averages their predictions (e.g., Bagging Classifier, Random Forest).




    -Boosting: Trains base models sequentially, where each new model tries to correct the errors of the previous ones (e.g., AdaBoost, Gradient Boosting, XGBoost).


    -Stacking (Stacked Generalization): Trains multiple diverse base models and then trains a meta-model (or "level-1" model) to combine their predictions.


14. What is ensemble learning in machine learning?

- Ensemble learning in machine learning is a meta-algorithm where multiple learning algorithms are used to obtain better predictive performance than could be obtained from any of the constituent learning algorithms alone. It involves combining the decisions from several individual models.

15. When should we avoid using ensemble methods?

- Ensemble methods might be avoided in situations where:

- Interpretability is paramount: Ensembles are often "black boxes" and can be difficult to interpret compared to single models.

- Computational resources are severely limited: Training multiple models can be computationally expensive and time-consuming.

- The dataset is very small: Ensemble methods might not offer significant advantages and could even overfit if the dataset is too small to create sufficiently diverse bootstrap samples.

- A simple, fast, and easily deployable model is required: The overhead of ensembles might not justify the marginal performance gain.

16. How does Bagging help in reducing overfitting?

- Bagging helps in reducing overfitting by reducing the variance of the model.  It achieves this by:


- Averaging/Voting: By aggregating predictions from multiple models trained on different subsets of data, Bagging smooths out the individual models' tendencies to overfit to specific patterns in their training subsets.

- Diversity: Bootstrap sampling ensures that each base model sees a slightly different version of the training data, leading to diverse models. This diversity prevents the ensemble from being overly reliant on any single model's biases.



17. Why is Random Forest better than a single Decision Tree?

- Random Forest is generally better than a single Decision Tree for several reasons:

- Reduced Overfitting: Single decision trees are prone to overfitting, especially deep ones. Random Forest reduces this by averaging the predictions of many trees, each trained on a different bootstrap sample and considering a random subset of features.

- Improved Accuracy: The combination of multiple diverse trees generally leads to higher predictive accuracy.

- Handles High Dimensionality: Random Forest can effectively handle datasets with a large number of features due to feature randomness.

- Robustness to Noise: The ensemble approach makes it more robust to noisy data.


18. What is the role of bootstrap sampling in Bagging?

- Bootstrap sampling is a crucial component of Bagging. Its role is to create multiple diverse subsets of the original training data.  Each subset is created by sampling with replacement from the original dataset, resulting in datasets of the same size as the original but with some samples repeated and some left out (OOB samples). This diversity in training sets allows the base models to learn different aspects of the data, which when aggregated, leads to a more robust and generalized ensemble.


19. What are some real-world applications of ensemble techniques?

- Ensemble techniques are widely used in various real-world applications, including:

- Medical Diagnosis: For predicting disease presence or patient outcomes.

- Fraud Detection: Identifying fraudulent transactions in financial systems.

- Image Classification: Object recognition and categorizing images.

- Credit Scoring: Assessing the creditworthiness of individuals.

- Weather Forecasting: Combining predictions from multiple models for more accurate forecasts.

- Customer Churn Prediction: Identifying customers likely to leave a service.

- Recommendation Systems: Providing personalized recommendations.


20. What is the difference between Bagging and Boosting?

- The main differences between Bagging and Boosting are:

- Parallel vs. Sequential: Bagging trains base models in parallel and independently. Boosting trains base models sequentially, where each new model learns from the errors of the previous ones.

- Variance vs. Bias Reduction: Bagging primarily aims to reduce variance by averaging predictions of diverse models. Boosting primarily aims to reduce bias by focusing on misclassified samples.

- Weighting: In Bagging, all base models typically have equal weight in the final prediction. In Boosting, base models (and data samples) are often weighted based on their performance, with better-performing models or harder-to-classify samples having more influence.

- Base Learner Strength: Bagging often uses "strong" base learners (e.g., deep decision trees). Boosting often uses "weak" base learners (e.g., shallow decision trees).

```python
# 21. Train a Bagging Classifier using Decision Trees on a sample dataset and print model accuracy.

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=10, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


base_estimator = DecisionTreeClassifier(random_state=42)



bagging_classifier = BaggingClassifier(estimator=base_estimator, n_estimators=100, random_state=42)


bagging_classifier.fit(X_train, y_train)


y_pred = bagging_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Bagging Classifier: {accuracy:.4f}")
```

```python
# 22. Train a Bagging Regressor using Decision Trees and evaluate using Mean Squared Error (MSE).

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np


X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, noise=0.5, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


base_estimator = DecisionTreeRegressor(random_state=42)


bagging_regressor = BaggingRegressor(estimator=base_estimator, n_estimators=100, random_state=42)


bagging_regressor.fit(X_train, y_train)


y_pred = bagging_regressor.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE) of Bagging Regressor: {mse:.4f}")
print(f"Root Mean Squared Error (RMSE) of Bagging Regressor: {np.sqrt(mse):.4f}")
```

```python
# 24. Train a Random Forest Regressor and compare its performance with a single Decision Tree.

from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np


X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, noise=10, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
y_pred_dt = dt_regressor.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
print(f"MSE of Single Decision Tree Regressor: {mse_dt:.4f}")
print(f"RMSE of Single Decision Tree Regressor: {np.sqrt(mse_dt):.4f}")


rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"MSE of Random Forest Regressor: {mse_rf:.4f}")
print(f"RMSE of Random Forest Regressor: {np.sqrt(mse_rf):.4f}")


if mse_rf < mse_dt:
    print("\nRandom Forest Regressor performed better (lower MSE) than the single Decision Tree.")
elif mse_dt < mse_rf:
    print("\nSingle Decision Tree Regressor performed better (lower MSE) in this specific run (less common for large datasets).")
else:
    print("\nBoth regressors performed similarly.")
```

```python
# 26. Train a Bagging Classifier using SVM as a base estimator and print accuracy.

from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler


wine = load_wine()
X = wine.data
y = wine.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


base_estimator = SVC(kernel='linear', random_state=42)



bagging_svm_classifier = BaggingClassifier(estimator=base_estimator, n_estimators=10, max_samples=0.7, random_state=42)


bagging_svm_classifier.fit(X_train_scaled, y_train)


y_pred = bagging_svm_classifier.predict(X_test_scaled)


accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Bagging Classifier with SVM base estimator: {accuracy:.4f}")
```

```python
# 27. Train a Random Forest Classifier with different numbers of trees and compare accuracy.

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt


cancer = load_breast_cancer()
X = cancer.data
y = cancer.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


n_estimators_list = [10, 50, 100, 200, 500]
accuracies = []

print("Comparing Random Forest accuracy with different numbers of trees:")
for n_estimators in n_estimators_list:


    rf_classifier = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    rf_classifier.fit(X_train, y_train)


    y_pred = rf_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"  N_Estimators = {n_estimators}: Accuracy = {accuracy:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(n_estimators_list, accuracies, marker='o', linestyle='-')
plt.title('Random Forest Accuracy vs. Number of Trees')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Accuracy')
plt.xticks(n_estimators_list)
plt.grid(True)
plt.show()
```

```python
# 30. Train an ensemble model using both Bagging and Random Forest and compare accuracy.

from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)



bagging_base_estimator = DecisionTreeClassifier(random_state=42)
bagging_classifier = BaggingClassifier(estimator=bagging_base_estimator, n_estimators=100, random_state=42, n_jobs=-1)
bagging_classifier.fit(X_train, y_train)
y_pred_bagging = bagging_classifier.predict(X_test)
accuracy_bagging = accuracy_score(y_test, y_pred_bagging)
print(f"Accuracy of Bagging Classifier: {accuracy_bagging:.4f}")



rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Accuracy of Random Forest Classifier: {accuracy_rf:.4f}")


if accuracy_rf > accuracy_bagging:
    print("\nRandom Forest Classifier performed better than the generic Bagging Classifier.")
elif accuracy_bagging > accuracy_rf:
    print("\nBagging Classifier performed better than Random Forest in this specific run (less common).")
else:
    print("\nBoth classifiers performed similarly.")
```

```python
# 31. Train a Random Forest Classifier and tune hyperparameters using GridSearchCV.

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score


cancer = load_breast_cancer()
X = cancer.data
y = cancer.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)



param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['sqrt', 'log2'],
    'max_depth': [None, 10, 20]
}


rf_classifier = RandomForestClassifier(random_state=42)


grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=1)


print("Starting GridSearchCV for Random Forest...")
grid_search.fit(X_train, y_train)
print("GridSearchCV finished.")


print("\nBest parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: {:.4f}".format(grid_search.best_score_))


best_rf_model = grid_search.best_estimator_


y_pred = best_rf_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test accuracy with best parameters: {test_accuracy:.4f}")
```

```python
# 32. Train a Bagging Regressor with different numbers of base estimators and compare performance.

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt


X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, noise=50, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


n_estimators_list = [5, 20, 50, 100, 200]
mse_scores = []

print("Comparing Bagging Regressor performance with different numbers of base estimators:")
for n_estimators in n_estimators_list:

    base_estimator = DecisionTreeRegressor(random_state=42)


    bagging_regressor = BaggingRegressor(estimator=base_estimator, n_estimators=n_estimators, random_state=42, n_jobs=-1)


    bagging_regressor.fit(X_train, y_train)


    y_pred = bagging_regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)
    print(f"  N_Estimators = {n_estimators}: MSE = {mse:.4f}")


plt.figure(figsize=(10, 6))
plt.plot(n_estimators_list, mse_scores, marker='o', linestyle='-')
plt.title('Bagging Regressor MSE vs. Number of Base Estimators')
plt.xlabel('Number of Base Estimators (n_estimators)')
plt.ylabel('Mean Squared Error (MSE)')
plt.xticks(n_estimators_list)
plt.grid(True)
plt.show()
```

```python
# 33. Train a Random Forest Classifier and analyze misclassified samples.

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import pandas as pd


cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
feature_names = cancer.feature_names
target_names = cancer.target_names


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)


y_pred = rf_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Overall Accuracy of Random Forest Classifier: {accuracy:.4f}")


misclassified_indices = np.where(y_test != y_pred)[0]
num_misclassified = len(misclassified_indices)
print(f"\nNumber of misclassified samples: {num_misclassified}")

if num_misclassified > 0:
    print("\nAnalyzing Misclassified Samples:")

    misclassified_data = pd.DataFrame(X_test[misclassified_indices], columns=feature_names)
    misclassified_data['True_Label'] = [target_names[i] for i in y_test[misclassified_indices]]
    misclassified_data['Predicted_Label'] = [target_names[i] for i in y_pred[misclassified_indices]]

    print(misclassified_data.head())



    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(pd.DataFrame(cm, index=target_names, columns=target_names))


    false_positives = np.where((y_test == 0) & (y_pred == 1))[0]

    false_negatives = np.where((y_test == 1) & (y_pred == 0))[0]

    print(f"\nNumber of False Positives (True Malignant, Predicted Benign): {len(false_positives)}")
    print(f"Number of False Negatives (True Benign, Predicted Malignant): {len(false_negatives)}")

else:
    print("No misclassified samples found on the test set.")
```

```python
# 36. Train a Stacking Classifier using Decision Trees, SVM, and Logistic Regression, and compare accuracy.

from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

wine = load_wine()
X = wine.data
y = wine.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

estimators = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svm', SVC(kernel='linear', probability=True, random_state=42)),
    ('lr', LogisticRegression(solver='liblinear', random_state=42))
]



final_estimator = LogisticRegression(solver='liblinear', random_state=42)


stacking_classifier = StackingClassifier(
    estimators=estimators,
    final_estimator=final_estimator,
    cv=5,
    n_jobs=-1
)


print("Training Stacking Classifier...")
stacking_classifier.fit(X_train_scaled, y_train)
print("Stacking Classifier training complete.")


y_pred_stacking = stacking_classifier.predict(X_test_scaled)


accuracy_stacking = accuracy_score(y_test, y_pred_stacking)
print(f"\nAccuracy of Stacking Classifier: {accuracy_stacking:.4f}")



dt_single = DecisionTreeClassifier(random_state=42)
dt_single.fit(X_train_scaled, y_train)
acc_dt = accuracy_score(y_test, dt_single.predict(X_test_scaled))
print(f"Accuracy of Single Decision Tree: {acc_dt:.4f}")


svm_single = SVC(kernel='linear', random_state=42)
svm_single.fit(X_train_scaled, y_train)
acc_svm = accuracy_score(y_test, svm_single.predict(X_test_scaled))
print(f"Accuracy of Single SVM: {acc_svm:.4f}")


lr_single = LogisticRegression(solver='liblinear', random_state=42)
lr_single.fit(X_train_scaled, y_train)
acc_lr = accuracy_score(y_test, lr_single.predict(X_test_scaled))
print(f"Accuracy of Single Logistic Regression: {acc_lr:.4f}")


print("\n--- Comparison ---")
print(f"Stacking Classifier Accuracy: {accuracy_stacking:.4f}")
print(f"Highest Single Model Accuracy: {max(acc_dt, acc_svm, acc_lr):.4f}")
if accuracy_stacking > max(acc_dt, acc_svm, acc_lr):
    print("Stacking Classifier performed better than any single base estimator.")
else:
    print("Stacking Classifier performance is comparable to or slightly lower than the best single base estimator.")
```

```python
# 40. Train a Bagging Regressor using different base estimators (DecisionTree and KNeighbors) and compare performance.

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np


X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, noise=50, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)



dt_base_estimator = DecisionTreeRegressor(random_state=42)
bagging_dt_regressor = BaggingRegressor(estimator=dt_base_estimator, n_estimators=100, random_state=42, n_jobs=-1)
bagging_dt_regressor.fit(X_train_scaled, y_train)
y_pred_dt_bagging = bagging_dt_regressor.predict(X_test_scaled)
mse_dt_bagging = mean_squared_error(y_test, y_pred_dt_bagging)
print(f"MSE of Bagging Regressor (Decision Tree Base): {mse_dt_bagging:.4f}")
print(f"RMSE of Bagging Regressor (Decision Tree Base): {np.sqrt(mse_dt_bagging):.4f}")



knn_base_estimator = KNeighborsRegressor(n_neighbors=5)
bagging_knn_regressor = BaggingRegressor(estimator=knn_base_estimator, n_estimators=100, random_state=42, n_jobs=-1)
bagging_knn_regressor.fit(X_train_scaled, y_train)
y_pred_knn_bagging = bagging_knn_regressor.predict(X_test_scaled)
mse_knn_bagging = mean_squared_error(y_test, y_pred_knn_bagging)
print(f"\nMSE of Bagging Regressor (KNeighbors Base): {mse_knn_bagging:.4f}")
print(f"RMSE of Bagging Regressor (KNeighbors Base): {np.sqrt(mse_knn_bagging):.4f}")


if mse_dt_bagging < mse_knn_bagging:
    print("\nBagging with Decision Trees performed better (lower MSE).")
elif mse_knn_bagging < mse_dt_bagging:
    print("\nBagging with KNeighborsRegressor performed better (lower MSE).")
else:
    print("\nBoth Bagging regressors performed similarly.")
```
